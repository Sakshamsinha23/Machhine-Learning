```python
data["Gender"].fillna(data["Gender"].mode,inplace=True)
```

```python
data["Married"].fillna(data["Married"].mode,inplace=True)
data.isnull().sum()
```

```python
# get dummies
# Convert 'Gender' and 'Married' columns to strings if needed
en_data = data[["Gender", "Married"]].astype(str)
pd.get_dummies(en_data).info()
```

**Scikit-learn library**

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

# Initialize the OneHotEncoder with drop='first'
ohe = OneHotEncoder( sparse_output=False)

# Fit and transform the data
arr = ohe.fit_transform(en_data)

# Get the correct column names
columns = ohe.get_feature_names_out(en_data.columns)

# Print the array and column names
print(arr)
print(columns)

# Create the DataFrame with the correct column names
encoded_df = pd.DataFrame(arr, columns=columns)

# View the result
print(encoded_df)
```

**Using Drop First Function**

```python
from sklearn.preprocessing import OneHotEncoder

ohe=OneHotEncoder(sparse_output=False,drop='first')

# fit and transform
arr = ohe.fit_transform(en_data)

columns=ohe.get_feature_names_out(en_data.columns)

encoded_df=pd.DataFrame(arr,columns=columns)
```
