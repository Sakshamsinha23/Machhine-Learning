```python
df=pd.DataFrame({
    "Size":['s','m','l','xl','s','m','l','s','s','l','xl','m']
})
df.head()
```

```python
ord_data=[['s','m','l','xl']]
ord_data
```

```python
#Using scikit learn

from sklearn.preprocessing import OrdinalEncoder

oe=OrdinalEncoder(categories=ord_data)
df['Size_en']= oe.fit_transform(df[['Size']])
df
```

**Using map function**

```python
ord_data1={'s':1,'m':2,'l':3,'xl':4}
df['Size_en1']=df['Size'].map(ord_data1)
df
```
