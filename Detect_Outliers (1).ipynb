**Detect outlier using IQR methodr**

```python
q1= data['CoapplicantIncome'].quantile(0.25)
q3= data['CoapplicantIncome'].quantile(0.75)
iqr=q3-q1
```

```python
min_range= q1 - (1.5 * iqr)
max_range= q3 + (1.5 * iqr)
```

```python
print("min range:",min_range)
print("max range:",max_range)
```

**REMOVE OUTLIER FROM ORIGINAL DATA**

```python
new_data =data[data['CoapplicantIncome'] <=max_range]
new_data.shape
```

```python
sns.boxplot(x='CoapplicantIncome',data=new_data)
plt.show()
```

**Removal of outlier using Z-Score**

```python
std= 3*data['CoapplicantIncome'].std()

min_range_z=data['CoapplicantIncome'].mean() - std
max_range_z=data['CoapplicantIncome'].mean() + std
```

```python
print(min_range_z, max_range_z)
```

```python

new_data_z = data[data['CoapplicantIncome'] <= max_range_z]
```

```python
new_data_z.shape
```

```python
sns.boxplot(x='CoapplicantIncome',data=new_data_z)
plt.show()
```
